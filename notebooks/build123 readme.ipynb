```python
from build123d import (
    Line,
    JernArc,
    PolarLine,
    make_face,
    make_hull,
)
from ocp_vscode import show

line = Line((0, -3), (6, -3))
line += JernArc(line @ 1, line % 1, radius=3, arc_size=180)
line += PolarLine(line @ 1, 6, direction=line % 1)
show(line)
```

```python
sketch = make_face(line)
show(sketch)
```
